```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv('bank-full.csv',sep=';')
```

```python
data
```

```python
d1 = pd.get_dummies(data,columns=['job','marital','education','contact','poutcome'])
```

```python
d1['default']=np.where(d1['default'].str.contains("yes"),1,0)
d1['housing']=np.where(d1['housing'].str.contains("yes"),1,0)
d1['loan']=np.where(d1['loan'].str.contains("yes"),1,0)
d1['y']=np.where(d1['y'].str.contains("yes"),1,0)
```

```python
d1
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
d1=d1.replace(order)
```

```python
d1
```

```python
d1.info()
```

```python
x=pd.concat([d1.iloc[:,0:11],d1.iloc[:,12:]],axis=1)
y=d1.iloc[:,11]
```

```python
lr=LogisticRegression()
lr.fit(x,y)
```

```python
y_pred=lr.predict(x)
y_pred
```

```python
y_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_df
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
lr.predict_proba(x)[:,1] 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fpr,tpr,thresholds=roc_curve(y,lr.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
